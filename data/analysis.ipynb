```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
old_excel = pd.read_excel('KnownSites_normalized.xlsx')

updated_sites = pd.read_excel('Updated_Sites.xlsx')
updated_sites.rename(columns={'Period':'Period_new'}, inplace=True)

display(old_excel.head())
display(updated_sites.head())
```

```python
comb =  pd.merge(old_excel, updated_sites, left_on='NAME', right_on='Site Name', how='inner', suffixes=('_old', '_new'))
comb.columns = comb.columns.str.title()
comb.head()
```

```python
for row in comb.iterrows():
    if not row[1]['Type Of Site'] == row[1]['Site Type']:
        print(row[1]['NAME'], row[1]['Type Of Site'], row[1]['Site Type'])
    if not row[1]['Period'] == row[1]['Period_New']:
        print(row[1]['Name'], row[1]['Period'], row[1]['Period_New'])
```

```python
sites = comb[['Name', 'Geographical Region', 'Geographical Location', 'Dd', 'Dms', 'Elevation (Masl)', 'Period_New', 'Site Type']]
sites.loc[:, 'Dd_ns'] = sites['Dd'].apply(lambda x: x.split(',')[0]).astype(float)
sites.loc[:, 'Dd_ew'] = sites['Dd'].apply(lambda x: x.split(',')[1]).astype(float)
sites
```

```python
sites.to_excel('UC_sites.xlsx', index=False)
```

```python
sites.to_csv('UC_sites.csv', index=False)
```
